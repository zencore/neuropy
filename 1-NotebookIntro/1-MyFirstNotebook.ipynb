## Basic python

We can use a "code" cell to run any valid python code.

For example, writing the following code (below) and hitting "run cell" will print the numbers 36 and 12.
If you change the numbers, and re-run, the output will change

````python
a = 12 
b = 24

c = a + b

print c
print a
````

If the **last line** of a cell creates any output, (or returns a value), it will be printed to the screen:

````python
d = 100
e = c + d
e
````

Note that this is only for the **last line**.
Also, remember that variable assignments does not result in an output:

```python
a = range(10)

```

will not create any output, whereas 
```python
range(10)
```

will. 

## Any python code is valid

You write **any** valid python code in the cells, including conditional statements, cycles, and even function definitions:

````python
for i in range(5):
    print i

for j in range(5):
    if j == 3:
        print "j is equal to 3"
    else:
        print "j is not equal to 3"

def silly_function(n):
    """ This silly function returns the word 'silly' 'n' number of times.
    """
    return "silly" * n

print silly_function(5)
````
