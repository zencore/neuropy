# Decoration

There are various ways we can customize the appearance of a plot in maplotlib. As always, don't forget to check the matplotlib documentation and gallery for information and inspiration:

- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
- http://matplotlib.org/gallery.html


The basic arguments you can pass to most plots in matplotlib are:

- *color* any matplotlib color
- *linestyle* [‘solid’ | ‘dashed’, ‘dashdot’, ‘dotted’ | (offset, on-off-dash-seq) | '-' | '--' | '-.' | ':' | 'None' | ' ' | '']
- *linewidth* float value in points
- *marker* A valid marker style
- *markeredgecolor* any matplotlib color
- *markeredgewidth*	float value in points
- *markerfacecolor* any matplotlib color
- *markersize* float



## Importing data

Well use the same dataset from last time: **eqs-years.txt**:

```python
handle = open("eqs-year.txt",'rU')

years = []
counts = []

for line in handle:
    cleaned_line = line.strip()
    data = cleaned_line.split(",")
    year = int(data[0])
    count = float(data[1])
    years.append(year)
    counts.append(count)

handle.close()
```

## Making the plot pretty

Lets try some of the arguments to make a pretty plot.

- the matplotlib list of colors: http://matplotlib.org/api/colors_api.html
- the maplotlib marker syles: http://matplotlib.org/api/markers_api.html

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(
    years, counts,
    markersize = 7,
    linewidth = 4,
    color='lightgray',
    markerfacecolor = 'gray',
    markeredgecolor='darkblue',
    markeredgewidth=1,
    marker='o'
    
)
plt.xlabel("Years")
plt.ylabel("Count")
plt.show()
```
