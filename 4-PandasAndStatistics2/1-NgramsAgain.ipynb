Just incase: a basic plot using matplotlib:


There are various ways we can customize the appearance of a plot in maplotlib. As always, don't forget to check the matplotlib documentation and gallery for information and inspiration:

- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
- http://matplotlib.org/gallery.html


The basic arguments you can pass to most plots in matplotlib are:

- *color* any matplotlib color
- *linestyle* [‘solid’ | ‘dashed’, ‘dashdot’, ‘dotted’ | (offset, on-off-dash-seq) | '-' | '--' | '-.' | ':' | 'None' | ' ' | '']
- *linewidth* float value in points
- *marker* A valid marker style
- *markeredgecolor* any matplotlib color
- *markeredgewidth*	float value in points
- *markerfacecolor* any matplotlib color
- *markersize* float

For example:

```python
plt.plot(x, y, color='red')      # makes a red line
```
